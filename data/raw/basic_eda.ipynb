```python
import pandas as pd

# 1. Load
path = "/content/drive/MyDrive/Data-MMDS/yellow_tripdata_2025-01.parquet"
df = pd.read_parquet(path)

# 2. Overview
print("OVERVIEW")
print("Shape (rows, cols):", df.shape)
print("\nColumns:", df.columns.tolist())

print("\nHEAD / TAIL")
print(df.head(5))
print("\nTail:")
print(df.tail(5))

# 3. Missing values
print("\nMISSING VALUES")
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df) * 100).round(2)
missing_table = pd.DataFrame({"missing": missing, "missing_%": missing_pct})
print(missing_table[missing_table["missing"] > 0].head(30))

# 4. Duplicates
print("\nDUPLICATES")
dup_rows = df.duplicated().sum()
print("Duplicate rows:", dup_rows)

# 5. Sample rows with any missing values
print("\nSAMPLE ROWS WITH MISSING VALUES")
rows_with_na = df[df.isna().any(axis=1)]
print("Rows with any NA:", len(rows_with_na))
if len(rows_with_na) > 0:
    print(rows_with_na.head(5))
```
