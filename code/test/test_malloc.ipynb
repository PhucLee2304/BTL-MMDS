```python
import pandas as pd
import os
import sys
import gc 
```

```python
def trigger_malloc_error_with_data():
    print("--- Starting Realistic Memory Allocation Test ---")

    # Get the absolute path of the current directory
    # In Jupyter, we can use os.getcwd() if the notebook is in code/test/
    current_dir = os.getcwd()

    # Move up 2 levels to BTL-MMDS root, then into data/raw/
    data_folder = os.path.join(current_dir, "../../data/raw/") 

    print(f"Data path in use: {os.path.abspath(data_folder)}")
    
    all_chunks = []
        
    try:
        iteration = 0
        while True:
            iteration += 1
            # Loading files from month 1 to 6
            for i in range(1, 7): 
                file_name = f"yellow_tripdata_2025-{i:02d}.parquet"
                file_path = os.path.join(data_folder, file_name)
                
                if os.path.exists(file_path):
                    # Load parquet file into RAM
                    # Note: User has experience with Python and Parquet files
                    df = pd.read_parquet(file_path)
                    all_chunks.append(df)
                    
                    # pd.concat creates a new copy, consuming significant RAM
                    full_df = pd.concat(all_chunks, ignore_index=True)
                    
                    # Calculate approximate DataFrame size in RAM
                    mem_usage = sys.getsizeof(full_df) / (1024**3) # Convert to GB
                    print(f"Iteration {iteration}, loaded {file_name}: {len(full_df):,} rows (~{mem_usage:.2f} GB)")
                else:
                    print(f"File not found: {file_name}")
                
    except MemoryError:
        print("\n[SUCCESS] MemoryError caught!")
        print("The system cannot allocate more memory for pd.concat.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
if __name__ == "__main__":
    trigger_malloc_error_with_data()
```
