```python
import pandas as pd
import os
import sys
import gc 
```

```python
def trigger_malloc_error_with_data():
    print("--- Starting Realistic Memory Allocation Test (2024 - 2025) ---")

    # Get current directory and locate the data/raw folder
    current_dir = os.getcwd()
    data_folder = os.path.join(current_dir, "../../data/raw/") 

    print(f"Data path in use: {os.path.abspath(data_folder)}")
    
    all_chunks = []
    years = [2024, 2025] # Sequential processing
        
    try:
        for year in years:
            print(f"\n--- Processing Year: {year} ---")
            for month in range(1, 13): # Full year coverage (1-12)
                file_name = f"yellow_tripdata_{year}-{month:02d}.parquet"
                file_path = os.path.join(data_folder, file_name)
                
                if os.path.exists(file_path):
                    # Load the file into a temporary DataFrame
                    df = pd.read_parquet(file_path)
                    all_chunks.append(df)
                    
                    # Consolidate data to test memory limits
                    # pd.concat is memory-intensive as it creates a new copy
                    full_df = pd.concat(all_chunks, ignore_index=True)
                    
                    # Monitor RAM consumption
                    mem_usage = sys.getsizeof(full_df) / (1024**3) # Convert to GB
                    print(f"Loaded {file_name}: {len(full_df):,} total rows (~{mem_usage:.2f} GB)")
                else:
                    print(f"Skipping: {file_name} (File not found)")
                    
        print("\n--- Finished processing all available data without a crash. ---")
                
    except MemoryError:
        print("\nMemoryError caught!")
        print("The system reached its allocation limit during pd.concat.")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")
```

```python
if __name__ == "__main__":
    trigger_malloc_error_with_data()
```
