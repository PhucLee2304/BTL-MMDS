```python
import pandas as pd
import os

print("--- Starting Realistic Memory Allocation Test (2015) ---")

# Get current directory and locate the data/raw folder
current_dir = os.getcwd()
data_folder = os.path.join(current_dir, "../../data/raw/") 

print(f"Data path in use: {os.path.abspath(data_folder)}")

all_chunks = []
years = [2015] # Sequential processing
data = pd.DataFrame()
for year in years:
    print(f"\n--- Processing Year: {year} ---")
    for month in range(1, 13): # Full year coverage (1-12)
        file_name = f"yellow_tripdata_{year}-{month:02d}.parquet"
        file_path = os.path.join(data_folder, file_name)
        
        if os.path.exists(file_path):
            all_chunks.append(pd.read_parquet(file_path))

        else:
            print(f"Skipping: {file_name} (File not found)")
                
data = pd.concat(all_chunks, ignore_index=True)
```
