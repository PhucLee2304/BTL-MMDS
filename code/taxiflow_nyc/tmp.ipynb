```python
import os
import requests

# 1. Định nghĩa thư mục lưu trữ (khớp với thư mục đẩy lên HDFS của bạn)
download_dir = "/workspace/data/raw"
os.makedirs(download_dir, exist_ok=True)

# 2. Base URL chứa dữ liệu Yellow Taxi của thành phố New York
base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{}-{:02d}.parquet"

years = [2025]
months = range(1, 13)

print(f"BẮT ĐẦU TẢI DỮ LIỆU TỪ 2020 ĐẾN 2023 VÀO: {download_dir}")
print("-" * 50)

# 3. Vòng lặp duyệt qua từng năm và từng tháng
for year in years:
    for month in months:
        # Tạo URL và tên file tương ứng
        url = base_url.format(year, month)
        filename = f"yellow_tripdata_{year}-{month:02d}.parquet"
        filepath = os.path.join(download_dir, filename)
        
        # Bỏ qua nếu file đã được tải thành công từ trước
        if os.path.exists(filepath):
            print(f"[-] File {filename} đã tồn tại trong ổ cứng, bỏ qua...")
            continue
            
        print(f"[+] Đang tải {filename}...")
        
        try:
            # Gửi request tải file, stream=True giúp tải file dung lượng lớn mà không tốn RAM
            response = requests.get(url, stream=True)
            
            if response.status_code == 200:
                with open(filepath, 'wb') as f:
                    # Ghi dữ liệu vào ổ cứng từng cụm 8KB
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                print(f"    => Thành công!")
            else:
                print(f"    => LỖI: Không tìm thấy file trên server (Mã lỗi {response.status_code}).")
                
        except Exception as e:
            print(f"    => LỖI KẾT NỐI: {e}")

print("-" * 50)
print("HOÀN TẤT QUÁ TRÌNH TẢI DỮ LIỆU!")
```
